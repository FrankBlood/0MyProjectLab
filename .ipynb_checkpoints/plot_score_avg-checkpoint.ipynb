```python
name_id = graphlab.SFrame.read_csv('name_id.txt', delimiter='\t', header=True)
```

```python
nl_stat = graphlab.SFrame.read_csv('nl_result.txt', delimiter=' ', header=False)
```

```python
nl_stat = nl_stat[nl_stat['X3']!=0]
```

```python
nl_stat.rename({'X1': 'user'})
nl_stat.rename({'X2': 'year'})
nl_stat.rename({'X3': 'num'})
nl_stat.rename({'X4': 'count'})
nl_stat.rename({'X5': 'f_score_every'})
nl_stat.rename({'X6': 'f_score'})
```

```python
nl_stat['user'].unique()
```

```python
name_id[name_id['id']=='48220291@N04']['name'][0]
```

```python
count = 1
for user in nl_stat['user'].unique():
    user_year = nl_stat[nl_stat['user'] == user]
    name = name_id[name_id['id'] == user]['name'][0]
    print user, name
    plt.figure(figsize=(10,50), dpi=200, facecolor="white")
    
    plt.subplot(10,1,count)
    plt.plot(user_year['year'], user_year['num'], color='m', marker="v", linewidth=1, label='count of every year')
    plt.plot(user_year['year'], user_year['count'], color='c', marker="o", linewidth=1, label='count')
    plt.title(name+'  year VS. count')
    plt.xlabel('year')
    plt.ylabel('count')
    plt.rcParams.update({'font.size': 10.5})
    plt.legend(loc='upper left', ncol=3, columnspacing=0.001, prop={'size':10})
    count += 1
```
