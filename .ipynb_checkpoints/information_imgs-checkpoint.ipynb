```python
import graphlab
```

```python
img_data = graphlab.SFrame.read_csv('img_datas.txt', delimiter='\t', header=False)
```

```python
import re
```

```python
def get_year(time):
    return re.split('-', re.split(' ', time)[0])[0]
```

```python
img_data['year'] = img_data['X3'].apply(get_year)
```

```python
year = img_data.groupby(key_columns='year', operations={'count': graphlab.aggregate.COUNT()})
```

```python
year = year[year['year'] != 'NULL']
```

```python
year = year.sort('year')
```

```python
year['year'] = year['year'].apply(int)
```

```python
year.sort('year', ascending=False)
```

```python
year_1799 = year[year['year']>1799]
```

```python
year_1799
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(year_1799['year'], year_1799['count'], color="c", linestyle="-", linewidth=1)
plt.title('year VS. count')
plt.xlabel('year')
plt.ylabel('count')
# plt.xscale('log')
plt.rcParams.update({'font.size': 10.5})
```
